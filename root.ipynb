```python
import pandas as pd
import numpy as np
```

```python
file_pbdb_path = "pbdb_data.csv"
```

```python
# Load the PBDB dataset
dino_pbdb_pd = pd.read_csv(file_pbdb_path)
dino_pbdb_pd.head()
```

```python
# Check the number of occurences of Spinosaurus
dino_pbdb_pd[dino_pbdb_pd['accepted_name'].str.startswith('Spinosaurus')]
```

```python
# A short description of the dataset
dino_pbdb_pd.head()
```

```python
# First we will clean the dataset
dino_pbdb_pd_clean = dino_pbdb_pd.drop_duplicates(subset=["accepted_name"])
dino_pbdb_pd_clean.fillna(dino_pbdb_pd_clean["life_habit"].mode()[0], inplace=True)
idx = dino_pbdb_pd_clean[dino_pbdb_pd_clean['accepted_name'].str.startswith('Spinosaurus')].index
dino_pbdb_pd_clean = dino_pbdb_pd_clean.drop(idx) # Remove Spinosaurus from the dataset for prediction
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 
from sklearn.neural_network import MLPClassifier
from sklearn import tree
import matplotlib.pyplot as plt

def encode_data(data):
    lb_make = LabelEncoder()
    return lb_make.fit_transform(data), lb_make
```

```python
# We get the labels, and encode categorical data
labels = dino_pbdb_pd_clean['life_habit']
y, label_encoder = encode_data(labels)
y
```

```python
# We get features, and encode categorical data
interval = dino_pbdb_pd_clean['early_interval']
f1,interval_encoder = encode_data(interval)
f1
```

```python
flags = dino_pbdb_pd_clean['flags']
f2, flags_encoder = encode_data(flags)
f2
```

```python
rank = dino_pbdb_pd_clean['accepted_rank']
f3, rank_encoder = encode_data(rank)
f3
```

```python
motility = dino_pbdb_pd_clean['motility']
f4, motility_encoder = encode_data(motility)
f4
```

```python
vision = dino_pbdb_pd_clean['vision']
f5, vision_encoder = encode_data(vision)
f5
```

```python
diet = dino_pbdb_pd_clean['diet']
f6, diet_encoder = encode_data(diet)
f6
```

```python
reproduction = dino_pbdb_pd_clean['reproduction']
f7, reproduction_encoder = encode_data(reproduction)
f7
```

```python
ontogeny = dino_pbdb_pd_clean['ontogeny']
f8, ontogeny_encoder = encode_data(ontogeny)
f8
```

```python
composition = dino_pbdb_pd_clean['composition']
f9, composition_encoder = encode_data(composition)
f9
```

```python
# Get the shapes of the filtered datasets
X = np.array([f1, f2, f3, f4, f5, f6, f7, f8, f9]).T
X.shape, y.shape
```

```python
# Our first prediction model will use Decision Trees
# Split training data into validation and training sets
# Perform 5-fold cross-validation

dt_best_model = None
dt_best_accuracy = -float('inf')
dt_validation_accuracies = []
dt_training_accuracies = []
for i in range(5):
    X_train, X_val, y_train, y_val = train_test_split(X, y, train_size=0.8, random_state=i)
    dt = tree.DecisionTreeClassifier()
    dt.fit(X_train, y_train)
    dt_y_hat_valid = dt.predict(X_val)
    dt_y_hat_train = dt.predict(X_train)
    dt_validation_accuracy = np.sum(dt_y_hat_valid == y_val) / len(y_val)
    dt_training_accuracy = np.sum(dt_y_hat_train == y_train) / len(y_train)
    dt_validation_accuracies.append(dt_validation_accuracy)
    dt_training_accuracies.append(dt_training_accuracy)
    if dt_validation_accuracy > dt_best_accuracy:
        dt_best_model = dt
        dt_best_accuracy = dt_validation_accuracy
dt_best_accuracy
```

```python
plt.figure(figsize=(12, 8))
plt.plot(dt_training_accuracies, label="Train Accuracy")
plt.plot(dt_validation_accuracies, label="Valid Accuracy")
plt.title("Decision Tree Model")
plt.xlabel("Cross-Validation #")
plt.xticks(np.arange(0,5,1))
plt.ylabel("Accuracy")
plt.show()
plt.savefig("tree_plot.png")
```

```python
sp_int = list(interval_encoder.transform(['Early Cenomanian']))
sp_fl = list(flags_encoder.transform(['epifaunal']))
sp_rank = list(rank_encoder.transform(['genus']))
sp_mot = list(motility_encoder.transform(['actively mobile']))
sp_vis = list(vision_encoder.transform(['epifaunal']))
sp_diet = list(diet_encoder.transform(['carnivore']))
sp_rep = list(reproduction_encoder.transform(['oviparous, dispersal=direct/internal,mobile']))
sp_ont = list(ontogeny_encoder.transform(['accretion,modification of parts']))
sp_comp = list(composition_encoder.transform(['hydroxyapatite']))
spinosaurus_data = np.array([sp_int, sp_fl, sp_rank, sp_mot, sp_vis, sp_diet, sp_rep, sp_ont, sp_comp]).flatten()
spinosaurus_data = spinosaurus_data.reshape(1, -1)
```

```python
sp_pred = dt_best_model.predict(spinosaurus_data)
label_encoder.inverse_transform(sp_pred)[0]
```

```python
# Our second prediction model will use Neural Networks
# Split training data into validation and training sets
# Perform 5-fold cross-validation

nn_best_model = None
nn_best_accuracy = -float('inf')
nn_validation_accuracies = []
nn_training_accuracies = []
for i in range(5):
    X_train, X_val, y_train, y_val = train_test_split(X,y, train_size=0.8, random_state=i)
    nn = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(30, 4), random_state=666)
    nn.fit(X_train, y_train)
    nn_y_hat_valid = nn.predict(X_val)
    nn_y_hat_train = nn.predict(X_train)
    nn_validation_accuracy = np.sum(nn_y_hat_valid == y_val) / len(y_val)
    nn_training_accuracy = np.sum(nn_y_hat_train == y_train) / len(y_train)
    nn_validation_accuracies.append(nn_validation_accuracy)
    nn_training_accuracies.append(nn_training_accuracy)
    if nn_validation_accuracy > nn_best_accuracy:
        nn_best_model = nn
        nn_best_accuracy = nn_validation_accuracy
nn_best_accuracy
```

```python
plt.figure(figsize=(12, 8))
plt.plot(nn_training_accuracies, label="Train Accuracy")
plt.plot(nn_validation_accuracies, label="Valid Accuracy")
plt.title("Neural Network Model")
plt.xlabel("Cross-Validation #")
plt.xticks(np.arange(0,5,1))
plt.ylabel("Accuracy")
plt.show()
plt.savefig("nn_plot.png")
```

```python
sp_preds = nn_best_model.predict_proba(spinosaurus_data)
# label_encoder.inverse_transform(sp_pred)
sp_pred_map = {}
for i in range(len(sp_preds[0])):    
    label = label_encoder.inverse_transform([i])[0]
    sp_pred_map[label] = sp_preds[0][i]
    
sp_pred_map = {k: v for k, v in sorted(sp_pred_map.items(), key=lambda item: -item[1])}
sp_pred_map
```

```python
plt.figure(figsize=(12, 8))
plt.bar([i+1 for i in range(len(sp_pred_map))], [sp_pred_map[k] for k in sp_pred_map.keys()], tick_label=list(sp_pred_map.keys()))
plt.title("Neural Network Probability Outputs")
plt.xlabel("Classification Probability")
plt.ylabel("Life Habitat")
plt.xticks(rotation=90)
plt.savefig("nn_prob_plot.png")
plt.show()

```
